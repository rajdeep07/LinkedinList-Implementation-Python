# Node class

```python
class Node():
      def __init__(self, val, next:'Node' = None):
        #NOTHING CAN BE ADDED
        self.val = val     #Note global. Slist can access
        self.next = next   #note global. Slist can access
```

# Slist class

```python
class Slist():
    def __init__(self):
        #NOTHING CAN BE ADDED
        self._first = None
        self._last = None
        
    #############################
    # All public functions
    #############################
    def append(self, data):
        if self._first is None:
            self._first = Node(data, None)
            return

        itr = self._first

        while itr.next:
            itr = itr.next

        itr.next = Node(data, None)
        
    # Function to insert node 
    def insert(self, root, item): 
        temp = Node(item) 

        if (root == None): 
            root = temp 
        else : 
            ptr = root 
            while (ptr.next != None): 
                ptr = ptr.next
            ptr.next = temp 

        return root
            
    def build_slist_from_list(self, arr):
        self._first = None
        self._last = None
        temp = arr
        if arr is not None:
            node = Node(val=arr[0])
            self._first = node
            for elem in arr[1:]:
                node.next = Node(val=elem)
                node = node.next
                
            self._last = node
            
               
    def __print__(self):
        if self._first is None:
            print("Linked list is empty")
            return
        itr = self._first
        llstr = ''
        while itr:
            llstr += str(itr.val)+' --> ' if itr.next else str(itr.val)
            itr = itr.next
        print(llstr)
        
    def __len__(self):
        count = 0
        itr = self._first
        while itr:
            count+=1
            itr = itr.next

        return count
            
    def prepend(self, data):
        node = Node(data, self._first)
        self._first = node
        
    def find(self, x): 
  
        current = self._first 
        while current != None: 
            if current.val == x: 
                return True #            
            current = current.next
            
        return False
    
    def delete(self, x): 
        
        temp = self._first 
        
        if (temp is not None): 
            if (temp.val == x): 
                self._first = temp.next
                temp = None
                return
 
        while(temp is not None): 
            if temp.val == x: 
                break
            prev = temp 
            temp = temp.next
 
        # if key was not present in linked list 
        if(temp == None): 
            return False
 
        # Unlink the node from linked list 
        prev.next = temp.next
 
        temp = None
    
    def is_empty(s):
        if len(s) == 0:
            return True
```
